```python
import os
#Permite realizar operaciones con la gestión de archivos, directorios, rutas, variables de entorno y otras funcionalidades del sistema operativo

def registrar_usuario():
  #Función para registrar un nuevo usuario en el sistema.
  #Args:
    #No recibe argumentos.
  #Returns:
    #No retorna ningún valor.
  

  nombre_usuario = input("Ingrese su nombre de usuario: ")
  contrasena = input("Ingrese su contraseña: ")

  # Verificar si el nombre de usuario ya existe
  if nombre_usuario in usuarios:
    print("El nombre de usuario ya está en uso. Intente nuevamente.")
    return

  # Almacenar la información del usuario en el diccionario y en el archivo
  usuarios[nombre_usuario] = contrasena
  guardar_usuarios()
  print("Usuario registrado exitosamente.")

def mostrar_usuarios():
  #Función para mostrar la información de todos los usuarios registrados.
  #Args:
    #No recibe argumentos.
  #Returns:
    #No retorna ningún valor.


  if not usuarios:
    print("No hay usuarios registrados.")
    return

  for usuario, contrasena in usuarios.items():
    print(f"Usuario: {usuario}, Contraseña: {contrasena}")

def iniciar_sesion():
  #Función para iniciar sesión en el sistema.
  #Args:
    #No recibe argumentos.
  #Returns:
    #True si el inicio de sesión es exitoso, False en caso contrario.

  nombre_usuario = input("Ingrese su nombre de usuario: ")
  contrasena_ingresada = input("Ingrese su contraseña: ")

  if nombre_usuario not in usuarios:
    print("El nombre de usuario no existe.")
    return False

  if usuarios[nombre_usuario] != contrasena_ingresada:
    print("Contraseña incorrecta.")
    return False

  print("Inicio de sesión exitoso.")
  return True

def cargar_usuarios():
  #Función para cargar la información de los usuarios desde el archivo de texto.
  #Args:
    #No recibe argumentos.
  #Returns:
    #No retorna ningún valor.

  global usuarios

  try:
    with open("usuarios.txt", "r") as archivo:
      for linea in archivo:
        usuario, contrasena = linea.strip().split(",")
        usuarios[usuario] = contrasena
  except FileNotFoundError:
    pass

def guardar_usuarios():
  #Función para guardar la información de los usuarios en el archivo de texto.
  #Args:
    #No recibe argumentos.
  #Returns:
    #No retorna ningún valor.

  with open("usuarios.txt", "w") as archivo:
    for usuario, contrasena in usuarios.items():
      archivo.write(f"{usuario},{contrasena}\n")



# Diccionario para almacenar la información de los usuarios
usuarios = {}

# Cargar la información de los usuarios al iniciar el programa
cargar_usuarios()


# Menú principal

while True:
  print("\nMenú principal:")
  print("1. Registrar usuario")
  print("2. Mostrar usuarios")
  print("3. Iniciar sesión")
  print("4. Salir")

  opcion = input("Ingrese una opción: ")

  if opcion == "1":
    registrar_usuario()
  elif opcion == "2":
    mostrar_usuarios()
  elif opcion == "3":
    if iniciar_sesion():
      # Opciones adicionales para el usuario logueado
      while True:
        print("\nMenú usuario logueado:")
        print("1. Cerrar sesión")
        print("2. Volver al menú principal")

        opcion_usuario = input("Ingrese una opción: ")

        if opcion_usuario == "1":
          break
        elif opcion_usuario == "2":
          break
        else:
          print("Opción inválida.")
  elif opcion == "4":
    print("Hasta Pronto.")
    break
  else:
    print("Opción inválida.")
```
